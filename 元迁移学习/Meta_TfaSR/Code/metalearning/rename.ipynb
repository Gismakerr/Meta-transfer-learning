```python
import os

def rename_tif_files(folder_path):
    for root, dirs, files in os.walk(folder_path):
        folder_name = os.path.basename(root)  # 当前子文件夹的名称
        for file in files:
            if file.lower().endswith('.tif'):
                old_path = os.path.join(root, file)
                # 替换文件名第一个字母
                new_filename = folder_name + file[1:]
                new_path = os.path.join(root, new_filename)

                # 重命名文件
                os.rename(old_path, new_path)
                print(f"Renamed: {file} -> {new_filename}")

# 使用方式：修改为你的根文件夹路径
folder_path = r'D:\元迁移学习\To_黄涛\元迁移学习\Meta_TfaSR\Code\metalearning\meta_data'
rename_tif_files(folder_path)
```
